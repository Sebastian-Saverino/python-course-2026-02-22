````python
checking = list([1,3.3,False,"clown"])
checking[0] #Indexing - Gives us the first number in the list (0 spot) 
checking[1:3] #Slicing - Begins at 1 spot and stop at the 3 spot
# print(checking) #Provides the entire list
checking[:] #Provides the entire list
my_dict = {
    'cookies':'dessert',
    'hummus':'condiemnt'
}
my_dict #calls your dictionary
my_dict.keys() #Returns all keys in the dictionary
my_dict["cookies"] #Returns the value for the key "cookies"
for value in my_dict.values():
    print(value)

for key in my_dict.keys():
    print(key)
````

````python
# Iterables - Objects that can be looped over


my_list = [1, 2, 3]
for x in my_list:
    print(x)




my_tuple = (1, 2, 3)
for x in my_tuple:
    print(x)




my_set = {1, 2, 3}

for x in my_set:
    print(x)




my_dict = {"a": 1, "b": 2}
for key in my_dict:
    print(key)

# or

for key, value in my_dict.items():
    print(key, value)



my_string = "data"
for char in my_string:
    print(char)



for i in range(5):
    print(i)
````
