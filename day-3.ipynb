```python
def plan_trip():
    # First we want to get the inputs
    destination = input("Enter your where you want to go: ")
    hotel_price_per_night = float(input("Enter hotel price per night: "))
    length_of_stay = int(input("Enter length of stay (nights): "))
    discount_percent = float(input("Enter discount percentage: "))

    # Calculate total before discount
    total_cost = hotel_price_per_night * length_of_stay

    # Calculate discount
    discount_amount = total_cost * (discount_percent / 100)

    # Final cost
    final_cost = total_cost - discount_amount

    # Print summary
    print("\nTrip Summary")
    print(f"Destination: {destination}")
    print(f"Length of Stay: {length_of_stay} nights")
    print(f"Hotel Price Per Night: ${hotel_price_per_night:.2f}")
    print(f"Discount Applied: {discount_percent}%")
    print(f"Total Trip Cost: ${final_cost:.2f}")


plan_trip()
```
