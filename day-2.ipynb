```python

# Student Code Review

# List of weekdays
# Lists in Python are zero-indexed, meaning counting starts at 0.
# So:
# "Sunday" = 0
# "Monday" = 1
# "Tuesday" = 2
# "Wednesday" = 3
# "Thursday" = 4
# "Friday" = 5
# "Saturday" = 6

weekdays = [
    'Sunday',
    'Monday',
    'Tuesday',
    'Wednesday',
    'Thursday',
    'Friday',
    'Saturday'
]

# The current day of the week that I chose
today = 'Saturday'

# A day of the week that I wish it was, but it is not
wish_it_was = 'Tuesday'


# Step 1: Find the index of "today"
# .index() searches through the list and returns
# the numeric position of the value inside the list.
#
# Example:
# weekdays.index("Friday") would return 5
#
# This value is already an integer, so we do not need to cast it with int().

today_index = weekdays.index(today)

# Step 2: Determine tomorrow
# If today's index is less than 6 (not Saturday),
# then tomorrow is simply the next position in the list.
#
# Since lists are zero-based:
# If today_index = 5 (Friday),
# then tomorrow would be at index 6 (Saturday).
#
# HOWEVER:
# If today_index = 6 (Saturday),
# we cannot add 1 because that would cause an IndexError.
# There is no index 7 in this list.
#
# So in that case, we wrap around back to index 0 (Sunday).

if today_index < 6:
    tomorrow = weekdays[today_index + 1]
else:
    tomorrow = weekdays[0]

# Step 3: Build the sentence
# The f in front of the string stands for "formatted string literal".
# It allows us to insert variables directly inside the string
# using curly braces {}.
#
# {today} will be replaced with the value of today.
# {tomorrow} will be replaced with the value of tomorrow.
# {wish_it_was} will be replaced with the value of wish_it_was.

sentence = f"Today is {today}. Tomorrow is {tomorrow}. I wish it was always {wish_it_was}."
# Step 4: Print the final result
# print() outputs the final formatted string to the console.

print(sentence)
```
